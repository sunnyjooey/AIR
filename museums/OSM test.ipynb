```python
import pandas as pd
```

```python
from IPython.display import display
```

```python
import osmnx as ox
```

```python
pd.set_option("display.max_columns", None)
```

```python
d2 = pd.read_csv('./IRS_ToLearnOnly_rev2.csv')
```

```python
d2
```

```python
import random
randomlist = random.sample(range(d2.shape[0]), 100)
```

```python
tags = {'tourism': True, 'leisure':True, 'historic':True}

e = []
w = {}

for i in randomlist:
    row = d2.loc[i, :]
    place = '{}, {}, {}, {}, USA'.format(row['street'], row['city'], row['state'], row['zip'])
    
    try:
        ret = ox.geometries.geometries_from_address(place, tags, dist=100)
        w[i] = ret
        
    except:
        e.append((i, row['name']))
```

```python
e
```

```python
for k,v in w.items():
    print(d2.loc[k, 'name'])
    display(v)
    print()
    print()
    
```

```python
for pair in e:
    try:
        res = ox.geocoder.geocode_to_gdf(pair[0])
        print(pair)
        display(res)
    except:
        pass
```
